### Código para transformar imágenes a .webp

```python
from PIL import Image
import os

# Carpeta donde están las imágenes PNG
input_folder = "imagenes_png"

# Carpeta de salida WEBP
output_folder = "imagenes_webp"

# Crear carpeta de salida si no existe
os.makedirs(output_folder, exist_ok=True)

# Recorrer archivos
for file_name in os.listdir(input_folder):

    # Solo archivos .png
    if file_name.lower().endswith(".png") or file_name.lower().endswith(".jpg") or file_name.lower().endswith(".jpeg"):

        png_path = os.path.join(input_folder, file_name)

        # Nombre salida
        webp_name = os.path.splitext(file_name)[0] + ".webp"
        webp_path = os.path.join(output_folder, webp_name)

        # Abrir imagen
        image = Image.open(png_path)

        # Convertir a RGB si tiene transparencia problemática
        if image.mode in ("RGBA", "P"):
            image = image.convert("RGBA")

        # Guardar como WEBP
        image.save(
            webp_path,
            "WEBP",
            quality=85,      # Calidad (0-100)
            method=6         # Mejor compresión
        )

        print(f"Convertida: {file_name} -> {webp_name}")

print("Conversión finalizada.")
```
